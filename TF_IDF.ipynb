```python
# Run all list will crash, but only first 5 rows can work. ture_news.head()
# True news tf-idf 
def tf_idf(articles_df):
  count_vectorizer = CountVectorizer()
  sparse_count_matrix = count_vectorizer.fit_transform(articles_df['text'])
  doc_term_matrix = sparse_count_matrix.todense()
  count_df = pd.DataFrame(doc_term_matrix, columns=count_vectorizer.get_feature_names())
  count_df.index = articles_df['title']

  transformer = TfidfTransformer()

  sparse_tf_idf_matrix = transformer.fit_transform(doc_term_matrix)

  tf_idf_matrix = sparse_tf_idf_matrix.todense()

  tf_idf_df = pd.DataFrame(tf_idf_matrix, columns=count_vectorizer.get_feature_names())

  tf_idf_df.index = articles_df['title']

  tf_idf_df_transposed = tf_idf_df.transpose(copy=True)
  return (tf_idf_df_transposed)
```

```python
# True news tf-idf
ture_news_tfidf = tf_idf(shuffle_True_df)
fake_news_tfidf = tf_idf(shuffle_Fake_df)
```

```python
top_x = 20
tf_idf_words_trueNews = []
for (columnName, columnData) in fake_news_tfidf.iteritems():
  word_series = columnData.sort_values(ascending=False)[:top_x]
  for index, value in word_series.items():
    tf_idf_words_trueNews.append(index)
pp = pprint.PrettyPrinter()
c = Counter(tf_idf_words_trueNews)
most_common_fake_news_tfidf = c.most_common(20)
print("20 MOST COMMON TF IDF WORDS IN THE Fake News")
pp.pprint(most_common_fake_news_tfidf)
```

# Topic Modelling

```python
shuffle_Fake_df.dropna(inplace=True)
shuffle_Fake_df["date"] = shuffle_Fake_df["date"].astype('datetime64[s]')
shuffle_Fake_df.shape
```

```python
shuffle_True_df.dropna(inplace=True)
shuffle_True_df["date"] = shuffle_True_df["date"].astype('datetime64[s]')
shuffle_True_df.shape
```

```python
def vectorization(text):
  tfidf_vectorizer = TfidfVectorizer(stop_words = 'english',
                                    lowercase = True,
                                    max_features = 8000,
                                    max_df = 0.95, 
                                    min_df = 5)

  dtm_tfidf = tfidf_vectorizer.fit_transform(text) 

  tfidf_feature_names = tfidf_vectorizer.get_feature_names()

  lda_tfidf = LatentDirichletAllocation(n_components=10, 
                                        max_iter=100,
                                        random_state=0)

  lda_tfidf.fit(dtm_tfidf)
  tfidf_topic_values = lda_tfidf.transform(dtm_tfidf)
  return lda_tfidf, tfidf_feature_names,tfidf_vectorizer,dtm_tfidf,tfidf_topic_values
```

```python
def display_topics(model, feature_names, num_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic {}:".format(topic_idx))
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-num_top_words - 1:-1]]))
```

```python
def topic_df(tfidf_topic_values):
  tfidf_topic_df = pd.DataFrame(tfidf_topic_values)

  tfidf_topic_df['top_topic'] = tfidf_topic_df.idxmax(axis=1)

  return tfidf_topic_df
```

```python
num_top_words = 20
lda_tfidf, tfidf_feature_names,tfidf_vectorizer,dtm_tfidf,tfidf_topic_values_fake = vectorization(shuffle_Fake_df["text"])
fake_news = display_topics(lda_tfidf, tfidf_feature_names, num_top_words)
fake_news
```
